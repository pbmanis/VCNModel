```python
from neuron import h

# using bokeh for graphics
from bokeh.io import output_notebook
import bokeh.plotting as plt
output_notebook()

# standard run library (gives h.continuerun)
h.load_file('stdrun.hoc')

# morphology
axon = h.Section(name='axon')
axon.L = 50000
axon.diam = 10
axon.nseg = 101

# biophysics
axon.insert('hh')

# current injection
ic = h.IClamp(axon(0))
ic.delay = 1
ic.dur = 1
ic.amp = 101

# setup a plot

# setup rangevarplot
rvp = h.RangeVarPlot('v')
rvp.begin(axon(0))
rvp.end(axon(1))
x = h.Vector()
y = h.Vector()
ys = []

def save_rvp_data():
    rvp.to_vector(y, x)
    ys.append(list(y))

# initialize and run, plotting twice
h.finitialize(-65)
h.continuerun(5)
save_rvp_data()
h.continuerun(10)
save_rvp_data()
h.continuerun(15)
save_rvp_data()

# show the plot
f = plt.figure(x_axis_label='position', y_axis_label='v')
f.multi_line([list(x)] * len(ys), ys, color=['black', 'red', 'blue'], line_width=2)
plt.show(f)
```
